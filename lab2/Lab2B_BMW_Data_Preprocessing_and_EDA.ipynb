# BMW Used Cars Dataset
## Basic Preprocessing and Exploratory Data Analysis

### 1. Import required libraries

```python

# Import essential libraries for data analysis and visualization
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

### 2. Load the dataset

```python
from google.colab import files

_ = files.upload()
```

```python

# Load the BMW dataset from the uploaded CSV file
df = pd.read_csv("bmw.csv")

# Display first 5 rows
df.head()
```

### 3. Basic dataset information

```python

# Check shape of the dataset
df.shape

# Check column names and data types
df.info()
```

### 4. Summary statistics

```python

# Generate basic statistical summary for numerical columns
df.describe()
```

### 5. Check for missing values

```python

# Check missing values in each column
df.isnull().sum()
```

### 6. Check for duplicate rows

```python

# Count duplicate rows
df.duplicated().sum()
```

### 7. Basic preprocessing

```python

# Convert column names to lowercase for consistency
df.columns = df.columns.str.lower()

# Verify updated column names
df.columns
```

### 8. Distribution of car prices

```python

# Visualize distribution of prices
plt.figure()
sns.histplot(df['price'], bins=30, kde=True)
plt.title("Distribution of BMW Car Prices")
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

### 9. Price vs Manufacturing Year

```python

# Scatter plot of price vs year
plt.figure()
sns.scatterplot(x=df['year'], y=df['price'])
plt.title("Price vs Manufacturing Year")
plt.xlabel("Year")
plt.ylabel("Price")
plt.show()
```

### 10. Mileage vs Price in USD

```python

# Scatter plot of mileage vs price
plt.figure()
sns.scatterplot(x=df['mileage'], y=df['price'])
plt.title("Mileage vs Price in USD")
plt.xlabel("Mileage")
plt.ylabel("Price in USD")
plt.show()
```

### 11. Fuel type distribution

```python

# Count plot for fuel types
plt.figure()
sns.countplot(x=df['fueltype'])
plt.title("Fuel Type Distribution")
plt.xlabel("Fuel Type")
plt.ylabel("Count")
plt.show()
```

### 12. Transmission type distribution

```python

# Count plot for transmission types
plt.figure()
sns.countplot(x=df['transmission'])
plt.title("Transmission Type Distribution")
plt.xlabel("Transmission")
plt.ylabel("Count")
plt.show()
```

### 13. Simple observations


- Newer BMW cars generally have higher prices.
- Cars with higher mileage tend to have lower prices.
- Petrol and diesel are the most common fuel types.
- Automatic transmission is more frequent in recent models.
