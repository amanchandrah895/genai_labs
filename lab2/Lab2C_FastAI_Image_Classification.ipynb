```python
from fastai.vision.all import *
```

```python
path = untar_data(URLs.PETS)
path
```

```python
path.ls()
```

```python
dls = ImageDataLoaders.from_name_func(
    path,
    get_image_files(path/'images'),
    valid_pct=0.2,
    seed=42,
    label_func=lambda x: x[0].isupper(),
    item_tfms=Resize(224)
)
```

```python
dls.show_batch(max_n=6)
```
